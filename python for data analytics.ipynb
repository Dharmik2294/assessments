```python
class ClinicAppointment:

    def __init__(self):
        self.appointments = {}
        self.slots = ["10am", "11am", "12pm", "2pm", "3pm"]

    def book_appointment(self):
        name = input("Enter patient name: ")
        age = int(input("Enter age: "))
        mobile = input("Enter mobile number: ")
        doctor = input("Preferred doctor: ")

        print("Available slots:", self.slots)
        slot = input("Choose slot: ")

        key = (doctor, slot)

        if key not in self.appointments:
            self.appointments[key] = []

        if len(self.appointments[key]) >= 3:
            print("Slot full!")
        else:
            self.appointments[key].append({
                "name": name,
                "age": age,
                "mobile": mobile
            })
            print("Appointment booked successfully!")

    def view_or_cancel(self):
        mobile = input("Enter mobile number: ")

        for key, patients in self.appointments.items():
            for p in patients:
                if p["mobile"] == mobile:
                    print("Appointment Found:", p, "Doctor:", key[0], "Slot:", key[1])
                    choice = input("Cancel appointment? (y/n): ")

                    if choice == "y":
                        patients.remove(p)
                        print("Appointment cancelled.")
                    return

        print("No appointment found.")
```

```python
cli = ClinicAppointment()
```

```python
cli.book_appointment()
```

```python
cli.view_or_cancel()
```

```python
cli.view_or_cancel()
```

```python
class SchoolManagement:

    def __init__(self):
        self.students = {}
        self.student_id = 1

    def new_admission(self):
        name = input("Enter name: ")
        age = int(input("Enter age: "))
        student_class = int(input("Enter class (1-12): "))
        mobile = input("Enter guardian mobile: ")

        if age < 5 or age > 18:
            print("Invalid age!")
            return

        if len(mobile) != 10:
            print("Invalid mobile number!")
            return

        sid = self.student_id

        self.students[sid] = {
            "name": name,
            "age": age,
            "class": student_class,
            "mobile": mobile
        }

        self.student_id += 1

        print("Student admitted. ID:", sid)

    def view_student(self):
        sid = int(input("Enter student ID: "))

        if sid in self.students:
            print(self.students[sid])
        else:
            print("Student not found")

    def update_student(self):
        sid = int(input("Enter student ID: "))

        if sid in self.students:
            mobile = input("New mobile: ")
            student_class = input("New class: ")

            self.students[sid]["mobile"] = mobile
            self.students[sid]["class"] = student_class

            print("Updated successfully")
        else:
            print("Student not found")

    def remove_student(self):
        sid = int(input("Enter student ID: "))

        if sid in self.students:
            del self.students[sid]
            print("Student removed")
        else:
            print("Student not found")
```

```python
std = SchoolManagement()
```

```python
std.new_admission()
```

```python
std.view_student()
```

```python
std.update_student()
```

```python
std.view_student()
```

```python
std.remove_student()
```

```python
std.view_student()
```

```python
import random

class BusReservation:

    def __init__(self):
        self.routes = {
            1: ("Mumbai to Pune", 500),
            2: ("Delhi to Jaipur", 600),
            3: ("Ahmedabad to Surat", 300)
        }

        self.tickets = {}
        self.seats = {1:40, 2:40, 3:40}

    def show_routes(self):
        for k,v in self.routes.items():
            print(k, v[0], "₹", v[1])

    def book_ticket(self):

        name = input("Enter passenger name: ")
        age = input("Enter age: ")
        mobile = input("Enter mobile: ")

        self.show_routes()
        route = int(input("Select route: "))

        if self.seats[route] == 0:
            print("No seats available")
            return

        ticket_id = random.randint(1000,9999)
        seat = 41 - self.seats[route]

        self.tickets[ticket_id] = {
            "name": name,
            "route": self.routes[route][0],
            "seat": seat
        }

        self.seats[route] -= 1

        print("Ticket booked!")
        print("Ticket ID:", ticket_id)

    def view_ticket(self):
        tid = int(input("Enter ticket ID: "))

        if tid in self.tickets:
            print(self.tickets[tid])
        else:
            print("Ticket not found")

    def cancel_ticket(self):
        tid = int(input("Enter ticket ID: "))

        if tid in self.tickets:
            del self.tickets[tid]
            print("Ticket cancelled")
        else:
            print("Ticket not found")
```

```python
bus = BusReservation()
```

```python
bus.book_ticket()
```

```python
bus.show_routes()
```

```python
bus.view_ticket()
```
