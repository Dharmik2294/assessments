```python
class ClinicAppointment:
    def __init__(self):
        self.slots = ["10am","11am","12pm","2pm","3pm"]
        self.appointments = []

    def book(self,name,age,mobile,doctor,slot):
        count = sum(1 for a in self.appointments
                    if a["doctor"]==doctor and a["slot"]==slot)
        if count >=3:
            return "Slot Full"
        self.appointments.append({
            "name":name,"age":age,"mobile":mobile,
            "doctor":doctor,"slot":slot})
        return "Appointment Booked"

    def view_or_cancel(self,mobile,cancel=False):
        for a in self.appointments:
            if a["mobile"]==mobile:
                if cancel:
                    self.appointments.remove(a)
                    return "Cancelled"
                return a
        return "Not Found"
```

```python
c=ClinicAppointment()
```

```python
c.book("Rahul",25,"9876543210","Dr Shah","10am")
```

```python
class SchoolManagement:
    def __init__(self):
        self.students = {}
        self.sid = 1001

    def admission(self,name,age,cls,mobile):
        if not (5<=age<=18):
            return "Invalid Age"
        if len(mobile)!=10:
            return "Invalid Mobile"

        sid = self.sid
        self.students[sid]={
            "name":name,"age":age,
            "class":cls,"mobile":mobile}
        self.sid+=1
        return sid

    def view(self,sid):
        return self.students.get(sid,"Not Found")

    def update(self,sid,cls=None,mobile=None):
        if sid in self.students:
            if cls:
                self.students[sid]["class"]=cls
            if mobile:
                self.students[sid]["mobile"]=mobile
            return "Updated"
        return "Not Found"

    def remove(self,sid):
        if sid in self.students:
            del self.students[sid]
            return "Removed"
        return "Not Found"
```

```python
s = SchoolManagement()
```

```python
s.admission("Rahul",15,10,"9876543210")
```

```python
import random

class BusReservation:
    def __init__(self):
        self.routes = {
            "Mumbai-Pune":500,
            "Delhi-Jaipur":600
        }
        self.tickets = {}
        self.max_seat = 40

    def book(self,name,age,mobile,route):
        seats = sum(1 for t in self.tickets.values()
                    if t["route"]==route)
        if seats>=self.max_seat:
            return "Bus Full"

        tid = random.randint(10000,99999)
        self.tickets[tid]={
            "name":name,"age":age,
            "mobile":mobile,"route":route,
            "seat":seats+1}
        return tid

    def view(self,tid):
        return self.tickets.get(tid,"Not Found")

    def cancel(self,tid):
        if tid in self.tickets:
            del self.tickets[tid]
            return "Cancelled"
        return "Not Found"
```

```python
b = BusReservation()
```

```python
b.book("Rahul",25,"9876543210","Mumbai-Pune")
```
